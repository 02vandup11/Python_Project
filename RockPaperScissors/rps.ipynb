```python
import random

def rps_game():
    """Rock-Paper-Scissors Game with Score System"""

    choices = ["rock", "paper", "scissors"]
    user_score = 0
    computer_score = 0

    print("\n  Rock-Paper-Scissors Game ")
    print("Enter 'q' to quit anytime.\n")

    while True:
        # User choice
        user_choice = input("Enter rock / paper / scissors: ").lower()
        
        # Quit option
        if user_choice == "q":
            print("\nExiting game...")
            break

        # Validate input
        if user_choice not in choices:
            print("Invalid choice! Try again.\n")
            continue

        # Computer choice
        computer_choice = random.choice(choices)

        print(f"You: {user_choice}")
        print(f"Computer: {computer_choice}")

        # Check winner
        if user_choice == computer_choice:
            print("Result: It's a DRAW!\n")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("Result:  You WIN!\n")
            user_score += 1
        else:
            print("Result: Computer WINS!\n")
            computer_score += 1

        # Print running scores
        print(f"üèÜ Score ‚Äî You: {user_score} | Computer: {computer_score}\n")

    # Final score
    print("\n Final Score ")
    print(f"You: {user_score} | Computer: {computer_score}")
    print("Goodbye! ")


if __name__ == "__main__":
    rps_game()
```
